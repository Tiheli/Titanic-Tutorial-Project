```python
# Load the Titanic dataset from the specified path and store it in the variable 'trainData'.
# Display the first three rows of the dataset to examine its structure and content.
trainData = pd.read_csv('/kaggle/input/titanic/train.csv')
trainData.head(3)
```

```python
# Load test dataset
testData = pd.read_csv('/kaggle/input/titanic/test.csv')
# Preview first 3 rows
testData.head(3)
```

```python
# Load the gender submission dataset (sample submission for predictions)
gender_submission = pd.read_csv('/kaggle/input/titanic/gender_submission.csv')
# Preview the dataset
gender_submission.head()
```

```python
# Filter rows where passengers are female and check their survival status
women = trainData.loc[trainData.Sex == 'female']["Survived"]
# Calculate the survival rate of women
rateWomen = sum(women) / len(women)
# Display the percentage of women who survived
print("% of women who survived:", rateWomen)
```

```python
# Filter rows where passengers are male and check their survival status
men = trainData.loc[trainData.Sex == 'male']["Survived"]

# Calculate the survival rate of men
rateMen = sum(men) / len(men)

# Display the percentage of men who survived
print("% of men who survived:", rateMen)
```

```python
# Importing the RandomForestClassifier for building a classification model
from sklearn.ensemble import RandomForestClassifier
```

```python
# Set the target variable
y = trainData["Survived"]

# Select features for prediction and convert categorical data into dummy variables
features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(trainData[features])  # Train data features
X_test = pd.get_dummies(testData[features])  # Test data features

# Initialize and train a RandomForestClassifier model
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)

# Make predictions on the test dataset
predictions = model.predict(X_test)

# Create a DataFrame for submission with PassengerId and predicted Survived values
output = pd.DataFrame({'PassengerId': testData.PassengerId, 'Survived': predictions})

# Save the predictions to a CSV file for submission
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

```python
# Combine predictions with the test data for display
result = pd.DataFrame({
    'PassengerId': testData['PassengerId'],  # Passenger ID from the test data
    'Predicted_Survived': predictions        # Predicted survival status (0 or 1)
})

# View the first 10 results
print(result.head(10))
```
